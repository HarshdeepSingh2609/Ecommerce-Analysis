Exploratory Data Analysis (EDA)


Purchase Price Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['Purchase Price'], bins=100, kde=True)
plt.title('Distribution of Purchase Price')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 5))
plt.boxplot(df['Purchase Price'])
plt.title('Distribution of Purchase Price')
plt.show()


```

```python


# Univariate analysis for categorical variables
categorical_variables = ['Language', 'Job', 'Browser Info', 'Company']
for variable in categorical_variables:
  value_counts = df[variable].value_counts().head(10)
  print(f"Frequency of {variable}:\n{value_counts}\n")
  plt.figure(figsize=(12, 6))
  plt.bar(value_counts.index, value_counts.values)
  plt.xlabel(variable)
  plt.ylabel('Frequency')
  plt.title(f'Bar Chart of {variable} Frequency')
  plt.xticks(rotation=90)
  plt.show()
```

BIVARIATE ANALYSIS

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

 Average Purchase Price by Credit Card Provider

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='CC Provider', y='Purchase Price', data=df, estimator='mean')
plt.title("Average Purchase Price by Credit Card Provider")
plt.xticks(rotation=45)
plt.show()
```

 Top Customers by Purchase Price

```python
top_customers = df.groupby('Email')['Purchase Price'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_customers.plot(kind='barh')
plt.title("Top 10 Customers by Total Purchase Price")
plt.xlabel("Total Purchase Price")
plt.ylabel("Customer Email")
plt.gca().invert_yaxis()
plt.show()
```

 Frequency of Purchases by Email Domain

```python
df['Domain']=df['Email'].str.split('@').str[1]
top_domains = df['Domain'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_domains.index, y=top_domains.values)
plt.title('Top 10 Email Domains by Number of Purchases')
plt.xlabel('Email Domain')
plt.ylabel('Purchase Count')
plt.xticks(rotation=45)
plt.show()
```

 Purchase Price by Job

```python




top_10_jobs = df['Job'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
df[df['Job'].isin(top_10_jobs)].boxplot(column='Purchase Price', by='Job', vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Distribution of Purchase Price Across Top 10 Jobs')
plt.ylabel('Purchase Price')
plt.xlabel('Job')
plt.xticks(rotation=45, ha='right')
plt.suptitle('')  # Remove default subplot title
plt.tight_layout()
plt.show()
```

 Purchase Price by Language

```python


plt.figure(figsize=(12, 6))
df.boxplot(column='Purchase Price', by='Language', vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Distribution of Purchase Price Across Languages')
plt.ylabel('Purchase Price')
plt.xlabel('Language')
plt.xticks(rotation=45, ha='right')
plt.suptitle('')
plt.tight_layout()
plt.show()
```
